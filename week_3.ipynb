```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 400)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin, color='green', linestyle='-', label='sin(x)')
plt.plot(x, y_cos, color='purple', linestyle='--', label='cos(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Curves')

plt.legend()

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers from a normal distribution
data = np.random.normal(0, 1, 1000)
mean_value=np.mean(data)

# Plot histogram
plt.hist(data,bins=20,color='red')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('normal distribution')
plt.axvline(mean_value,label='mean')
plt.legend()
plt.show()
```

```python
import pandas as pd

# File path
file_path = '../penguins(in).csv'

# Reading the CSV file
data = pd.read_csv(file_path)

# Displaying the first few rows of the dataframe
print(data.head(10))
print(data.info())
mean_length=data['bill_length_mm'].mean()
median_length=data['bill_length_mm'].median()
std_length=data['bill_length_mm'].std()
print(mean_length, median_length, std_length)
type_species=data['species'].value_counts()
print(type_species)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = '../penguins(in).csv'

# Reading the CSV file
data = pd.read_csv(file_path)

body_mass = data['body_mass_g']
flipper_length = data['flipper_length_mm']
fig,ax=plt.subplots()
ax.plot(body_mass,flipper_length,label='Body Mass')
ax.set_xlabel('Body Mass')
ax.set_ylabel('Flipper Length')
ax.set_title('Body Mass/Flipper Length')
ax.grid(True)
plt.show()
```
