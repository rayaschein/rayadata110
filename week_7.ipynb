```python
import pandas as pd

data=pd.read_csv('penguins(in).csv')
data.head(10)
```

```python
data.describe()
```

```python
filtered=data[data['body_mass_g'] >4000]
filtered.head(5)
```

```python
filter_by_bill_length = data[(data['bill_length_mm'] > 50) & (data['bill_length_mm'] < 55) & data['bill_depth_mm'] < 18]
filter_by_bill_length.shape
```

```python
average_flipper_length=data.groupby(['species','island'])['flipper_length_mm'].mean().sort_values(ascending=False)
average_flipper_length
```

```python
import matplotlib.pyplot as plt

grouped=data.groupby('species')['body_mass_g']
plt.figure(figsize=(8, 6))
grouped.plot(kind='hist', bins=20, alpha=0.6, legend=True)
plt.show()
#ask professor, why only show gentoo, and question
```

```python
colors = ['red', 'yellow', 'blue']
unique_islands = data['island'].unique()

for island, color in zip(unique_islands, colors):
  subset = data[data['island'] == island]
  plt.hist(subset['flipper_length_mm'], bins=10, alpha=0.5, color=color, label=island)
  plt.xlabel('Flipper Length (mm)')
  plt.ylabel('Frequency')
  plt.title('Distribution of Flipper Length by Island')
  plt.legend()
  plt.show()
```
