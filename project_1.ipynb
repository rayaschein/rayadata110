```python
import pandas as pd

file_path = 'surgery_timing_THS (2).csv'

data = pd.read_csv(file_path)
data['gender'] = data['gender'].replace({0.0: 'Female', 1.0: 'Male'})
data.dropna(subset=['gender'], inplace=True)
data.head(10)
```

```python
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(data=data, x='gender', y='ccsMort30Rate', hue='gender')
plt.xlabel('Gender')
plt.ylabel('Mortality Risk')
plt.title('Mortality by Gender')
plt.subplot(1, 2, 2)
sns.barplot(data=data, x='gender', y='complication', hue='gender')
plt.xlabel('Gender')
plt.ylabel('Complication Risk')
plt.title('Complication Risk by Gender')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='hour', y='ccsComplicationRate', color='blue')
sns.regplot(data=data, x='hour', y='ccsComplicationRate', scatter=False, color='red')
plt.xlabel('Hour')
plt.ylabel('Complication Risk')
plt.title('Surgery time vs. Complication Risk')
plt.tight_layout()
plt.show()
```

```python
sns.lmplot(data=data, x='age', y='ccsComplicationRate', line_kws={'color': 'orange'}, aspect=1.5, height=6)
plt.xlabel('Age')
plt.ylabel('Complication Risk')
plt.title('Age vs. Complication Risk')
sns.lmplot(data=data, x='bmi', y='ccsComplicationRate', line_kws={'color': 'orange'}, aspect=1.5, height=6)
plt.xlabel('BMI')
plt.ylabel('Complication Risk')
plt.title('BMI vs. Complication Risk')
plt.show()
```
